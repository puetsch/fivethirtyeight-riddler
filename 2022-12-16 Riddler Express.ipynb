# Tribonacci to 2023

https://fivethirtyeight.com/features/can-you-make-it-to-2023/

```python
NMAX = 100

for i in range(NMAX+1):
    for j in range(i,NMAX+1):
        for k in range(j,NMAX+1):
#             print(f"Testing sequence {i,j,k}...")
            l = 0
            seq = [i,j,k]
            while True:
                s = sum(seq[-3:])
                seq.append(s)
                if s == 0: break
                l += 1
                if s == 2023:
                    print(f"** {i+j+k:3}: after {l:3} iterations: {seq}")
                if s >= 2023:
#                     print(seq)
                    break
```
